## üìà Tourism Growth Over Time

Azerbaijan's tourism industry has shown remarkable growth from 2001 to 2024.




```python
import plotly.express as px
import plotly.graph_objects as go

# Line chart - Overall trend
fig = px.line(df, x='Year', y='Total_Persons', 
              title='Total Hotel Guests in Azerbaijan (2001-2024)',
              markers=True)

fig.update_traces(
    line_color='#3498db', 
    line_width=3, 
    marker=dict(size=10, color='#e74c3c')
)

fig.update_layout(
    template='plotly_white',
    height=500,
    xaxis_title='Year',
    yaxis_title='Number of Guests',
    hovermode='x unified',
    font=dict(size=13)
)

fig.show()

# Key statistics
print("\nüìä KEY GROWTH METRICS:")
print("="*60)
print(f"2001: {df[df['Year']==2001]['Total_Persons'].values[0]:,.0f} guests")
print(f"2024: {df[df['Year']==2024]['Total_Persons'].values[0]:,.0f} guests")

growth_2001_2024 = ((df[df['Year']==2024]['Total_Persons'].values[0] / 
                     df[df['Year']==2001]['Total_Persons'].values[0]) - 1) * 100
print(f"Total Growth (2001-2024): {growth_2001_2024:.1f}%")

# Pre-COVID
print(f"\nPre-COVID (2019): {df[df['Year']==2019]['Total_Persons'].values[0]:,.0f} guests")
# COVID
print(f"COVID Low (2020): {df[df['Year']==2020]['Total_Persons'].values[0]:,.0f} guests")
covid_drop = ((df[df['Year']==2020]['Total_Persons'].values[0] / 
               df[df['Year']==2019]['Total_Persons'].values[0]) - 1) * 100
print(f"COVID Impact: {covid_drop:.1f}%")

# Recovery
recovery_2024 = ((df[df['Year']==2024]['Total_Persons'].values[0] / 
                  df[df['Year']==2019]['Total_Persons'].values[0]) - 1) * 100
print(f"Recovery (2024 vs 2019): {recovery_2024:+.1f}%")
```

### üåç Citizens vs Foreign Tourists

How has the composition of tourism changed over the years?


```python
# Stacked area chart
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df['Year'], 
    y=df['Citizens_Persons'],
    name='Citizens',
    mode='lines',
    stackgroup='one',
    fillcolor='rgba(52, 152, 219, 0.7)',
    line=dict(width=0.5, color='rgb(52, 152, 219)')
))

fig.add_trace(go.Scatter(
    x=df['Year'], 
    y=df['Foreigners_Persons'],
    name='Foreign Tourists',
    mode='lines',
    stackgroup='one',
    fillcolor='rgba(231, 76, 60, 0.7)',
    line=dict(width=0.5, color='rgb(231, 76, 60)')
))

fig.update_layout(
    title='Citizens vs Foreign Tourists (2001-2024)',
    xaxis_title='Year',
    yaxis_title='Number of Persons',
    template='plotly_white',
    height=500,
    hovermode='x unified'
)

fig.show()

# Foreign tourist growth
print("\nüåç FOREIGN TOURIST ANALYSIS:")
print("="*60)
foreign_2001 = df[df['Year']==2001]['Foreigners_Persons'].values[0]
foreign_2024 = df[df['Year']==2024]['Foreigners_Persons'].values[0]
print(f"2001: {foreign_2001:,.0f} foreign tourists")
print(f"2024: {foreign_2024:,.0f} foreign tourists")
print(f"Growth: {((foreign_2024/foreign_2001)-1)*100:.1f}%")
```

## ü¶† COVID-19 Pandemic Impact

The pandemic dramatically affected tourism. Let's analyze the recovery.


```python
# Year-over-year growth rate
df['Growth_Rate'] = df['Total_Persons'].pct_change() * 100

# Create period labels
df['Period'] = df['Year'].apply(lambda x: 
    'Pre-COVID' if x < 2020 
    else 'COVID Era' if x <= 2021 
    else 'Recovery'
)

# Bar chart
fig = px.bar(df, x='Year', y='Growth_Rate',
             title='Year-over-Year Tourism Growth Rate (%)',
             color='Period',
             color_discrete_map={
                 'Pre-COVID': '#2ecc71',
                 'COVID Era': '#e74c3c',
                 'Recovery': '#3498db'
             })

fig.add_hline(y=0, line_dash="dash", line_color="gray", line_width=2)

fig.update_layout(
    template='plotly_white',
    height=500,
    xaxis_title='Year',
    yaxis_title='Growth Rate (%)',
    showlegend=True
)

fig.show()
```
